## Use this notebook to schedule daily behavioral experiments based on number of animals, number/length of trials and desired start time.

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
import numpy as np
import random

from custom_modules.my_functions import create_schedule, generate_gantt_chart
```

```python

# Input from the user
num_animals = int(input("Enter the number of animals: "))
start_time = input("Enter the start time (hh:mm AM/PM): ")
num_trials = int(input("Enter the number of trials: "))

# Input trial durations as a list
trial_durations = []
for trial in range(num_trials):
    duration = int(input(f"Enter duration for Trial {trial + 1} (in minutes): "))
    trial_durations.append(duration)

# Input the intertrial interval
time_between_trials = int(input("Enter the time between each animal (in minutes): "))

# Create the schedule and display the visual timeline
schedule_df, timeline_data = create_schedule(num_animals, start_time, num_trials, trial_durations, time_between_trials)

# Display the schedule
print("\nExperiment Schedule:")
print(schedule_df)

# Generate and display the Gantt chart
generate_gantt_chart(timeline_data, num_animals)
```
